```python
import numpy as np
import matplotlib.pyplot as plt


# Years
years = np.array([1981, 1983, 1985, 1987, 1989, 1991, 1993, 1995, 1997, 1999])
# Temperatures
temperatures = np.array([14.1999, 14.2411, 14.0342, 14.2696, 14.197, 14.3055, 14.1853, 14.3577, 14.4187, 14.3438])

# Even years and corresponding temperatures
even_years = [1982, 1984, 1986, 1988, 1990, 1992, 1994, 1996, 1998]

linear_interpolation = [14.2205, 14.13765, 14.1519, 14.2333, 14.25125, 14.2454, 14.2715, 14.3882, 14.38125]
quadratic_interpolation = [14.2515, 14.1686, 14.0966, 14.2718, 14.2286, 14.2739, 14.2349, 14.4021, 14.3982]
cubic_interpolation = [14.1580, 14.1823, 14.2066, 14.2309, 14.2552, 14.2795, 14.3038, 14.3282, 14.3525]
regression = [14.2946, 14.1255, 14.1435, 14.2412, 14.2542, 14.2414, 14.2601, 14.4036, 14.3997]

# Plotting data sesuai hasil hitungan
plt.plot(even_years, linear_interpolation, label ='Linear Interpolation')
plt.plot(even_years, quadratic_interpolation, label ='Quadratic Interpolation')
plt.plot(even_years, cubic_interpolation, label ='Cubic Interpolation')
plt.plot(even_years, regression, label ='Regression Square')

plt.scatter(years, temperatures, label ='Odd Years')
plt.xlabel('Years')
plt.ylabel('Temperature (C)')
plt.title('Relationship Between Temperature and Year')
plt.legend()
plt.show()
```

```python
import sympy as sp
import math

#NOMOR 2 A

x = sp.symbols('x')

#Ekspansi Taylor orde ke empat sin x
f = sp.sin(x)
sin_taylor_series = sp.series(f, x, 0, 5).removeO() 

#Ekspansi Taylor orde ke empat cos x
f = sp.cos(x)
cos_taylor_series = sp.series(f, x, 0, 5).removeO()

#Ekspansi Taylor orde ke empat sin x cos x
f = sp.sin(x) * sp.cos(x)
sin_cos_taylor_series = sp.series(f, x, 0, 5).removeO()
    
print("Ekspansi Taylor Sin x :", sin_taylor_series)
print("Ekspansi Taylor Cos x :", cos_taylor_series)
print("Ekspansi Taylor Sin x Cos x :", sin_cos_taylor_series)

#Nomor 2 C

# Evaluasi cos(/4)
x = math.pi/4
cos_approximation = cos_taylor_series.subs('x', x)

# Truncation error bound
fifth_order_term = (x**5)/math.factorial(5)
error_bound = abs(fifth_order_term)

print("Aproksimasi cos(/4):", cos_approximation)
print("Error Bound:", error_bound)
```
